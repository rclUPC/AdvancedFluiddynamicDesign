```python
import numpy as np
```

```python
h = 0.3
phi_m = 0.5
psi_m = 0.3
DF = 0.5
N = 40
```

```python
rm = 0.5*(1+h)
Rm = 1-psi_m/2
alpha2_m = np.rad2deg(np.arctan(psi_m/phi_m))
beta1_m = np.rad2deg(np.arctan(1/phi_m))
beta2_m = np.rad2deg(np.arctan((1-psi_m)/phi_m))
tlr_m = 2.0/psi_m*(np.sqrt(phi_m*phi_m+np.square(1.0-psi_m))-(1.0-DF)*np.sqrt(1+phi_m*phi_m))
tls_m = 2.0/psi_m*(phi_m-(1.0-DF)*np.sqrt(phi_m*phi_m+psi_m*psi_m))
```

```python
r = np.linspace(h,1,N)
rbyrm = r*2/(1+h)
```

```python
phi = phi_m/rbyrm
psi = psi_m/rbyrm**2
R = 1 - psi/2
beta1 = np.rad2deg(np.arctan(1/phi))
beta2 = np.rad2deg(np.arctan((1-psi)/phi))
alpha2 = np.rad2deg(np.arctan(psi/phi))
tlr = 2.0/psi*(np.sqrt(phi*phi+np.square(1.0-psi))-(1.0-DF)*np.sqrt(1+phi*phi))
tls = 2.0/psi*(phi-(1.0-DF)*np.sqrt(phi*phi+psi*psi))
```

```python
import pandas as pd
```

```python
data = np.array([r,phi,psi,R,beta1,beta2,tlr,alpha2,tls]).transpose()
columns = ['r','phi','psi','R','beta1','beta2','t/l_R','alpha2','t/l_S']
fan = pd.DataFrame(data,columns=columns)
```
